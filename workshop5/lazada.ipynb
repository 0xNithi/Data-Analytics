```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CSV_PATH = "./lazada_txn_clean.csv"
df = pd.read_csv(CSV_PATH)
```

```python
dfg = df.groupby(['cust_name_hash']).agg(
    {'paid_price': 'sum', 'lazada_sku': 'count'}).sort_values(by='paid_price', ascending=False)
```

```python
# check data exploration all
sns.pairplot(dfg, x_vars=dfg.columns, y_vars=dfg.columns)
```

```python
dfg
```

```python
# scatter plot
dfg.plot.scatter(x='paid_price', y='lazada_sku', c='DarkBlue')
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(dfg)

print(kmeans.labels_)

dfg["clust"] = kmeans.labels_
dfg
```

```python
colors = ["red", "blue", "green", 'yellow', 'violet']

dfg.plot.scatter(x="paid_price", y="lazada_sku", c=dfg["clust"].astype(int).apply(lambda x: colors[x]))
```

```python
_k = []
_sse = []

for k in range(1, 10):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(dfg)
  _k.append(k)
  _sse.append(kmeans.inertia_)

sse_df = pd.DataFrame({"k":_k, "sse":_sse})
sse_df.plot(x="k", y="sse")
```
